```python
import torch
from torchvision import models, transforms
from tqdm import tqdm
```

```python
device = torch.device("cpu")

# Load pretrained ResNet-50 (include top minus final fc)
resnet = models.resnet50(pretrained=True)
resnet.fc = torch.nn.Identity()
resnet.eval().to(device)

# Preprocessing pipeline
preprocess = transforms.Compose([
    transforms.Resize((256, 128)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

def extract_feat(img_path):
    img = Image.open(img_path).convert("RGB")
    img_t = preprocess(img).unsqueeze(0).to(device)
    with torch.no_grad():
        f = resnet(img_t).cpu().numpy().squeeze()
    return f

# Paths
crop_root = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Output")
feat_dir = Path(r"C:\Users\ABHINAV\Desktop\player-reid\TrackletEmbeddings")
feat_dir.mkdir(parents=True, exist_ok=True)

# Loop over tracklets
for track_folder in tqdm(list(crop_root.iterdir())):
    if not track_folder.is_dir(): continue
    feats = []
    for img in sorted(track_folder.glob("*.jpg")):
        feats.append(extract_feat(img))
    if feats:
        avg = np.mean(np.vstack(feats), axis=0)
        np.save(feat_dir / f"{track_folder.name}.npy", avg)
```

```python
import os
import torch
import numpy as np
from pathlib import Path
from PIL import Image
from tqdm import tqdm
import torchvision.transforms as T
import torch.nn.functional as F
from torchreid.utils import FeatureExtractor

# -------------------------------
# 1. Function to Extract Embeddings
# -------------------------------
def extract_embeddings_from_crops(crop_root_folder, extractor, transform, output_path):
    crop_root_folder = Path(crop_root_folder)
    output_path = Path(output_path)
    output_path.parent.mkdir(parents=True, exist_ok=True)

    extractor.model.eval()
    extractor.model.to("cpu")

    tracklet_embeddings = {}

    for tracklet_folder in tqdm(sorted(crop_root_folder.iterdir()), desc="Processing tracklets"):
        if not tracklet_folder.is_dir():
            continue
        track_id = tracklet_folder.name
        embeddings = []

        for img_path in sorted(tracklet_folder.glob("*.jpg")):
            try:
                img = Image.open(img_path).convert("RGB")
                img_tensor = transform(img).unsqueeze(0).to("cpu")
                with torch.no_grad():
                    feat = extractor(img_tensor)
                embeddings.append(feat.cpu().numpy())
            except Exception as e:
                print(f"❌ Failed on {img_path.name}: {e}")

        if embeddings:
            avg_embedding = np.mean(np.vstack(embeddings), axis=0)
            tracklet_embeddings[track_id] = avg_embedding

    torch.save(tracklet_embeddings, output_path)
    print(f"\n✅ Saved embeddings to: {output_path}")

# -------------------------------
# 2. Load Model and Define Transform
# -------------------------------
extractor = FeatureExtractor(
    model_name='resnet50',
    model_path='',
    device='cpu'
)

transform = T.Compose([
    T.Resize((256, 128)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# -------------------------------
# 3. Extract Broadcast Embeddings
# -------------------------------
broadcast_crop_dir = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Output\broadcast")
broadcast_embed_path = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Embeddings\broadcast_embed.pt")

extract_embeddings_from_crops(broadcast_crop_dir, extractor, transform, broadcast_embed_path)

# -------------------------------
# 4. Matching Embeddings
# -------------------------------
# Load both embedding dictionaries
BROADCAST_EMBED_PATH = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Embeddings\broadcast_embed.pt")
TACTICAM_EMBED_PATH = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Embeddings\tacticam_embed.pt")

broadcast_embeddings = torch.load(BROADCAST_EMBED_PATH, weights_only=False)
tacticam_embeddings = torch.load(TACTICAM_EMBED_PATH, weights_only=False)

# Convert numpy arrays to tensors
broadcast_embeddings = {k: torch.from_numpy(v) if isinstance(v, np.ndarray) else v
                        for k, v in broadcast_embeddings.items()}
tacticam_embeddings = {k: torch.from_numpy(v) if isinstance(v, np.ndarray) else v
                       for k, v in tacticam_embeddings.items()}

# Matching: cosine similarity
matches = {}  # {broadcast_id: tacticam_id}
for b_id, b_embed in broadcast_embeddings.items():
    b_embed = b_embed.unsqueeze(0)  # [1, D]
    best_sim = -1
    best_tact_id = None

    for t_id, t_embed in tacticam_embeddings.items():
        t_embed = t_embed.unsqueeze(0)  # [1, D]
        sim = F.cosine_similarity(b_embed, t_embed).item()

        if sim > best_sim:
            best_sim = sim
            best_tact_id = t_id

    matches[b_id] = best_tact_id

# -------------------------------
# 5. Print Matches
# -------------------------------
print("\n📋 Broadcast to Tacticam Matches:")
for b_id, t_id in matches.items():
    print(f"Broadcast ID {b_id} → Tacticam ID {t_id}")
```

```python
import torch
import numpy as np
from pathlib import Path
import csv

# Load embeddings
BROADCAST_EMBED_PATH = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Embeddings\broadcast_embed.pt")
TACTICAM_EMBED_PATH = Path(r"C:\Users\ABHINAV\Desktop\player-reid\Embeddings\tacticam_embed.pt")

broadcast_embeddings = torch.load(BROADCAST_EMBED_PATH, weights_only=False)
tacticam_embeddings = torch.load(TACTICAM_EMBED_PATH, weights_only=False)

# Convert numpy arrays to torch tensors (if needed)
for k, v in broadcast_embeddings.items():
    if isinstance(v, np.ndarray):
        broadcast_embeddings[k] = torch.from_numpy(v)

for k, v in tacticam_embeddings.items():
    if isinstance(v, np.ndarray):
        tacticam_embeddings[k] = torch.from_numpy(v)

# Match each broadcast track to the closest tacticam track
broadcast_to_tacticam_matches = {}
for b_id, b_embed in broadcast_embeddings.items():
    b_embed = b_embed.unsqueeze(0)  # Shape (1, D)
    min_dist = float('inf')
    matched_t_id = None
    for t_id, t_embed in tacticam_embeddings.items():
        t_embed = t_embed.unsqueeze(0)  # Shape (1, D)
        dist = torch.nn.functional.pairwise_distance(b_embed, t_embed).item()
        if dist < min_dist:
            min_dist = dist
            matched_t_id = t_id
    broadcast_to_tacticam_matches[b_id] = matched_t_id

# Save matches to CSV
output_csv_path = Path(r"C:\Users\ABHINAV\Desktop\player-reid\matches.csv")

with open(output_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Broadcast_ID", "Tacticam_ID"])  # Header
    for b_id, t_id in broadcast_to_tacticam_matches.items():
        writer.writerow([b_id, t_id])

print(f"✅ Matches saved to: {output_csv_path}")
```
